```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_clusters_vs_locations(summary_csv_path: str) -> None:
    """
    Load the summary CSV file and create a plot for the number of clusters versus delivery locations.
    
    Parameters:
    - summary_csv_path (str): Path to the summary CSV file containing the clustering statistics.
    
    Returns:
    - None
    """
    # Load the summary CSV file
    summary_df = pd.read_csv(summary_csv_path)

    # Extract delivery locations and mean number of clusters
    delivery_locations = summary_df['Num_Delivery_Locations']
    mean_clusters = summary_df['mean']

    # Create the plot
    plt.figure(figsize=(12, 6))
    
    # Plot mean number of clusters for each delivery location count
    plt.plot(delivery_locations, mean_clusters, marker='o', markersize=8, linewidth=2, color='teal', alpha=0.8)
    
    # Fill area under the curve for better visualization
    plt.fill_between(delivery_locations, mean_clusters, color='lightseagreen', alpha=0.2)

    # Set labels, title, and legend
    plt.xlabel("Number of Delivery Locations", fontsize=14)
    plt.ylabel("Average Number of Clusters", fontsize=14)
    plt.title("Average Number of Clusters vs Delivery Locations", fontsize=16)
    plt.xticks(delivery_locations, fontsize=12)
    plt.yticks(fontsize=12)
    plt.grid(visible=True, which='both', linestyle='--', linewidth=0.7, alpha=0.7)

    # Remove top and right spines for a cleaner look
    plt.gca().spines['top'].set_visible(False)
    plt.gca().spines['right'].set_visible(False)

    # Show the plot
    plt.show()



# Example usage
plot_clusters_vs_locations('summary.csv')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_clusters_vs_locations(summary_csv_path: str) -> None:

    summary_df = pd.read_csv(summary_csv_path)

    # Extract delivery locations and number of clusters
    delivery_locations = summary_df['Num_Delivery_Locations']
    cluster_data = [summary_df.loc[loc, 'min':'max'].values for loc in summary_df.index]

    # Create the box plot
    plt.figure(figsize=(12, 6), dpi=300)
    
    # Box plot for each delivery location count
    plt.boxplot(cluster_data, labels=delivery_locations, patch_artist=True, boxprops=dict(facecolor='lightseagreen', color='teal'), 
                medianprops=dict(color='red'), whiskerprops=dict(color='teal'), capprops=dict(color='teal'))

    # Set labels, title, and legend
    plt.xlabel("Number of Delivery Locations", fontsize=14)
    plt.ylabel("Number of Clusters", fontsize=14)
    #plt.title("Number of Clusters vs Delivery Locations", fontsize=16)
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)
    plt.grid(visible=True, which='both', linestyle='--', linewidth=0.7, alpha=0.7)

    # Remove top and right spines for a cleaner look
    plt.gca().spines['top'].set_visible(False)
    plt.gca().spines['right'].set_visible(False)

    # Show the plot
    plt.tight_layout
    plt.show()

# Example usage
plot_clusters_vs_locations('summary.csv')
```
